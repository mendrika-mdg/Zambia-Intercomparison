# Number of cores per context domain for June-September (2004-2024)

```python
import numpy as np                                                                      
import matplotlib.pyplot as plt                                                         
```

```python
nb_cores = np.load("/home/users/mendrika/Zambia-Intercomparison/outputs/nbx0/nbx0-zambia.npy")
```

# Long lead time

```python
plt.figure(figsize=(7, 5))
plt.hist(nb_cores, bins=20, color="green", alpha=0.7, edgecolor='black')
plt.axvline(x=3, color='red', linestyle='--', linewidth=2, label=f'Mean={round(np.mean(nb_cores))}')
plt.xlabel("Number of Cores", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

# Save high-resolution figure
plt.savefig("core_distribution_300.png", dpi=300)
plt.show()
```
